### Scraping data of consumption by type of alcoholic beverage data corresponding to the year 2019 

```python
# Importing Libraries.  
from bs4 import BeautifulSoup
import requests
```

```python
# specifying the url of the webpage.
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_alcohol_consumption_per_capita'
# retrieving response object in a page.
page = requests.get(url)
# pulling the info from above url
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
# finding all tables. 
soup.find_all('table')
```

```python
# finding the elements of the requierd table. 
soup.find_all('table')[2]
```

```python
soup.find('table', class_ = 'wikitable sortable static-row-numbers')
```

```python
table = soup.find_all('table')[2]
print(table)
```

```python
# Retreiving only the table titles
table_titles = table.find_all('th')
table_titles
```

```python
# Comprehending a list 
table_title_list = [title.text.strip() for title in table_titles]
print (table_title_list)
```

```python
# Creating a data frame and inserting the data 
import pandas as pd
```

```python
df = pd.DataFrame(columns=table_title_list)
df
```

```python
# retreiving the values to be inserted in the rows (checking all rows).
column_data = table.find_all('tr')
```

```python
# Looping the info with values.
for row in column_data[1:]:
    row_data = row.find_all('td')
    row_data_list = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = row_data_list
    
```

```python
# Calling the final data frame.
df
```
